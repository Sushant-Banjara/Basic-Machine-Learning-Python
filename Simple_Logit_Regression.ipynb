```python
#Import the data
df1 = pd.read_csv('bank-data.csv')
df2 = pd.read_csv('bank-data-testing.csv')
print('The number of columns in training and test data are '+str(len(df1)) +' and '+str(len(df2))+' respectively')
```

```python
#concatenate the data and check the length
df = pd.concat([df1, df2])
print(len(df))
df.head()
```

```python
#Map the response variable to 0 and 1
#add a constant
df['y']= df['y'].map({'no':0, 'yes':1})
df = sm.add_constant(df)
df.head()
```

```python
#Split into training and testing set
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size =0.2)
print('The number of columns in training and test data are '+str(len(train)) +' and '+str(len(test))+' respectively')
train.head()
```

```python
#Separate response and explanatory variables
Y = train['y']
X = train.drop(columns=['Unnamed: 0', 'y'])
X.head()
```

```python
logreg = sm.Logit(Y,X).fit()
logreg.summary()
```

```python
#Predict and create confusion matrix for test data
test = test.drop(columns=['Unnamed: 0'])
test.head()
```

```python
pred = logreg.predict(test.drop(columns=['y']))
pred.head()
```

```python
bins=np.array([0,0.5,1]) 
cm = np.histogram2d(test['y'], pred, bins=bins)[0]
cm
```

```python
new_pred=[]
new_pred=[]
for i in pred:
    if i < 0.5:
        new_pred.append(0)
    else:
        new_pred.append(1)
        
```

```python
from sklearn.metrics import confusion_matrix as cm
matrix = cm (test['y'], new_pred)
matrix
```
